Join tables "country_language" and "country", filter only official languages

```python
pd.read_sql(
"""
SELECT 
  *
FROM
  country_language cl
INNER JOIN
  country c
ON
  cl.country_code = c.code
WHERE cl.is_official = 'True'
""",
"postgresql://netology:NetoSQL2019@84.201.153.170:19001/world-db"
)
```

Do the same work using Pandas. First, merge df_country_language and merge df_country

```python
combined_df = pd.merge(df_country_language, df_country, how='inner', left_on='country_code', right_on='code')
```
